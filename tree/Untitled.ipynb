```python
import dendropy
import collections
```

```python
treelist = dendropy.TreeList()
treelist.read_from_path("mcr.masked.afa.boottrees", schema="newick")

splits = []
for tree in treelist:
    tree_splits = [split.split_bitmask for split in tree.encode_bipartitions()]
    splits += tree_splits
    
split_counts = collections.Counter(splits)
split_props = {k: v/len(treelist) for k,v in split_counts.items()}
```
